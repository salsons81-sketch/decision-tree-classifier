```python
import pandas as pd
import os

# Check if 'fruits.csv' exists, if not, create a dummy one
if not os.path.exists('fruits.csv'):
    # Create dummy data
    dummy_data = {
        'color': ['red', 'green', 'yellow', 'red', 'green'],
        'shape': ['round', 'round', 'oval', 'round', 'oval'],
        'fruit': ['apple', 'apple', 'banana', 'strawberry', 'grape']
    }
    dummy_df = pd.DataFrame(dummy_data)
    dummy_df.to_csv('fruits.csv', index=False)
    print("Created a dummy 'fruits.csv' file as it was not found.")

data = pd.read_csv('fruits.csv')
print(data)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
```

```python
X=pd.get_dummies(data[['color', 'shape']])
y=data['fruit']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

model= DecisionTreeClassifier()
model.fit(X_train,y_train )

prediction  = model.predict(X_test)
print(f"Model prediction: {prediction}")
print(f"Actual Answers: \n{y_test}")

score = model.score(X_test, y_test)
print(f"Accuracy: {score}")
```
